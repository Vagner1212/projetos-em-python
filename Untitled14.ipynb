```python
#Vizualizando meus dados
import plotly.express as px

# Criar o histograma
fig = px.histogram(df, x="loja", y="preco", text_auto=True, color="forma_pagamento")
fig.show()
fig.write_html("histograma.html")
```

```python
fig = px.histogram(df, x="estado", y="preco", text_auto=True, color="forma_pagamento")
fig.show()
```

```python

# Dataframe com as colunas 'loja', 'ano_mes', and 'preco'

dados = df.groupby(["loja", "ano_mes"])["preco"].sum().reset_index()
dados["acumulado"] = dados.groupby("loja")["preco"].cumsum()

fig = px.bar(dados,
             x="acumulado",
             y='loja',
             color="loja",
             text="acumulado",
             range_x=[0, 123000],
             animation_frame="ano_mes")

fig.show()
```
